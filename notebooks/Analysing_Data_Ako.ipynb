```python
#Q1.Which platform has the highest average revenue per game? 

rev_per_game  = cleaned_df_n.groupby(['Platform', 'Game Title'])['Revenue (Millions $)'].mean()
rev_per_game_df = rev_per_game.reset_index()
print(type(rev_per_game_df))
rev_per_game_df.sort_values(by='Revenue (Millions $)', ascending=False) 
rev_per_game_df.head(100)

mean_revenue_sorted = (
    rev_per_game_df.groupby('Platform')['Revenue (Millions $)']
    .mean()
    .sort_values(ascending=False)
)

print(mean_revenue_sorted)

mean_revenue_sorted_df = mean_revenue_sorted.reset_index()
print(type(mean_revenue_sorted_df))
display(mean_revenue_sorted_df) 
```

```python
#which has the highest median revenue per game
median_revenue_sorted = (
    rev_per_game_df.groupby('Platform')['Revenue (Millions $)']
    .median()
    .sort_values(ascending=False)
)

print(median_revenue_sorted)

median_revenue_sorted_df = median_revenue_sorted.reset_index()
print(type(median_revenue_sorted_df))
display(median_revenue_sorted_df) 
```

```python
#cleaned_df_n is the source of df 
df['release_year'] = pd.to_datetime(df['release_year'], unit='ns')
df['release_year'] = df['release_year'].dt.year.astype('Int64')

df_2022_2024 = df[(df["release_year"] >= 2022) & (df["release_year"] <= 2024)]
df_2022_2024

print("--- Check for Missing Values ---")
print(df['release_year'].isna().sum()) 

print("\n--- Value Counts of Converted Years ---")
print(df['release_year'].value_counts(dropna=False).sort_index())

# Create a boolean mask for the valid range
year_mask = (df["release_year"] >= 2020) & (df["release_year"] <= 2024)

df_2020_2024 = df[year_mask]


# Assuming you have a clean copy available:
df = df.copy() 


df['release_year'] = pd.to_numeric(
    df['release_year'], 
    errors='coerce'
).astype('Int64') 

# Now the filtering will work
cleaned_df_n_2020_2024 = df[
    (df["release_year"] >= 2020) & 
    (df["release_year"] <= 2024)
].copy()

print(f"Number of rows after filtering 2022-2024: {len(df_2022_2024)}")

df_2020_2024
```

```python
# we sort to order the revenues and calculate the growth
# calculate the total revenue per year for each company
revenue_by_year = (
    df_2020_2024.groupby(["developer", "release_year"])["revenue_millions"]
           .sum()
           .reset_index()
           .sort_values(["developer", "release_year"])
)

# 2. Calculate year after year growth (in %) 
revenue_by_year["revenue_growth_%"] = (
    revenue_by_year.groupby("developer")["revenue_millions"]
                   .pct_change() * 100
)

print(revenue_by_year)
```

```python
#calculating revenue growth 2020-2024 
import pandas as pd

# Define the list of developers to keep
selected_developers = [
    'square enix', 
    'microsoft', 
    'ubisoft', 
    'bandai namco', 
    'rockstar', 
    'activision'
]

df_selected_six =  df_2020_2024[
     df_2020_2024['Developer'].isin(selected_developers)
].copy()

revenue_by_six = (
    df_selected_six.groupby(["developer", "release_year"])["revenue_millions"]
    .sum()
    .reset_index()
    .sort_values(["developer", "release_year"])
)

# 2. Calculate year after year growth (in %)
revenue_by_six["revenue_growth_%"] = (
    revenue_by_six.groupby("developer")["revenue_millions"]
    .pct_change() * 100
)

print("--- Data Filtered to Selected Six Developers (with Growth Rate) ---")
print(revenue_by_six)
```

```python
#line graph of revenue growth 
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
for company, group in revenue_by_six.groupby("Developer"):
    plt.plot(group["Release Year"], group["revenue_growth_%"], marker='o', label=company)

plt.title("Revenue Growth Rate (%) Over Time by Company")
plt.xlabel("Year")
plt.ylabel("Growth (%)")
plt.axhline(0, colors = ['#5f8195', '#5f8195','','#5f8195', '#5f8195', '#5f8195'], linestyle='--')  # ligne neutre
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the list of colors you want to use.
# You specified 6 colors (based on your previous list length of 6):
# ['#5f8195', '#5f8195','#dd3d6e','#5f8195', '#5f8195', '#5f8195']
# Note: I'll use a slightly more diverse set of colors for better visibility, 
# assuming you wanted to highlight one specific company (e.g., the red/pink one).
# If you want the exact colors you listed, you can replace this list.

# Assuming the developer names in revenue_by_six are, for example:
# 'Activision', 'Bandai Namco', 'Microsoft', 'Rockstar', 'Square Enix', 'Ubisoft'
# Matplotlib plots them based on the sorted keys from groupby.

# Get the list of developer names in the order they will be plotted (alphabetical)
# We assume 'revenue_by_six' contains the data.
developers_in_order = sorted(revenue_by_six['Developer'].unique())

# Define the colors. The length MUST match the number of developers (6).
# I'll use 6 distinct colors for maximum clarity:
custom_colors = [
    '#9e9e9e',  # Teal
    '#6abfda',  # Pink (This highlights one company well, like your red one)
    '#dd3e6e',  # Light Blue
    '#111111',  # Orange
    '#270901',  # Green
    '#212121'   # Red-Orange '#6abfda', '#9e9e9e',
]

# Create a mapping of Developer name to color
color_map = dict(zip(developers_in_order, custom_colors))
print(f"Color Mapping: {color_map}")

# --- Plotting ---

plt.figure(figsize=(12, 6))

# Use the color map inside the loop
for company, group in revenue_by_six.groupby("Developer"):
    plt.plot(
        group["Release Year"], 
        group["revenue_growth_%"], 
        marker='o', 
        label=company,
        color=color_map[company] # Use the specific color for this company
    )

plt.title("Revenue Growth Rate (%) Over Time by Top Growth Rate Developers", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Growth (%)", fontsize=12)

# Corrected axhline for the neutral line
plt.axhline(0, color='gray', linestyle='--') 

# Placing the legend outside for better line visibility
plt.legend(title='Developer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
