```python
import pandas as pd 
import yaml

try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("Configuration file not found!")

config

df = pd.read_csv(config['input_data'] ['file']) 
df
```

```python
cleaned_df = df.copy()
```

```python
#cleaning datetime 
def datetime (): 
 cleaned_df["Release Year"] = pd.to_datetime(cleaned_df["Release Year"],errors="coerce")
 return cleaned_df 
#df["Release Year"] = pd.to_numeric(df["Release Year"], errors='coerce')
#df["Release Year"] = df["Release Year"].astype("Int64")
datetime () 
```

```python
#reset index 
def index_reset () : 
    cleaned_df.reset_index(drop=True, inplace=True)

    cleaned_df.index = cleaned_df.index + 1
    cleaned_df.index.name = None

    return cleaned_df 
    
index_reset ()
```
