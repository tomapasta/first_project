```python
#Q1.Which platform has the highest average revenue per game? 

rev_per_game  = cleaned_df_n.groupby(['Platform', 'Game Title'])['Revenue (Millions $)'].mean()
rev_per_game_df = rev_per_game.reset_index()
print(type(rev_per_game_df))
rev_per_game_df.sort_values(by='Revenue (Millions $)', ascending=False) 
rev_per_game_df.head(100)

mean_revenue_sorted = (
    rev_per_game_df.groupby('Platform')['Revenue (Millions $)']
    .mean()
    .sort_values(ascending=False)
)

print(mean_revenue_sorted)

mean_revenue_sorted_df = mean_revenue_sorted.reset_index()
print(type(mean_revenue_sorted_df))
display(mean_revenue_sorted_df) 
```

```python
#which has the highest median revenue per game
median_revenue_sorted = (
    rev_per_game_df.groupby('Platform')['Revenue (Millions $)']
    .median()
    .sort_values(ascending=False)
)

print(median_revenue_sorted)

median_revenue_sorted_df = median_revenue_sorted.reset_index()
print(type(median_revenue_sorted_df))
display(median_revenue_sorted_df) 
```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



METASCORE_COLUMN = 'Metacritic Score'


cleaned_df_n[METASCORE_COLUMN] = pd.to_numeric(cleaned_df_n[METASCORE_COLUMN], errors='coerce').astype('Int64')
cleaned_df_n['Release Year'] = pd.to_numeric(cleaned_df_n['Release Year'], errors='coerce').astype('Int64')

cleaned_df_n.dropna(subset=[METASCORE_COLUMN, 'Release Year'], inplace=True)

START_YEAR = 2000
END_YEAR = 2024

df_filtered = cleaned_df_n[
    (cleaned_df_n['Release Year'] >= START_YEAR) & 
    (cleaned_df_n['Release Year'] <= END_YEAR)
].copy()


top_platforms_releases = df_filtered['Platform'].value_counts().nlargest(5).index.tolist()
print(f"\n By Most Active Platforms (by releases): {top_platforms_releases}")

df_top5 = df_filtered[df_filtered['Platform'].isin(top_platforms_releases)].copy()

metascore_pivot = df_top5.pivot_table(
    index='Release Year',
    columns='Platform',
    values=METASCORE_COLUMN,
    aggfunc='mean'
)

metascore_long = metascore_pivot.stack().reset_index(name='Average_Metacritic_Score')
metascore_long.rename(columns={'level_0': 'Release Year', 'level_1': 'Platform'}, inplace=True)

print("\n--- Aggregated Metacritic Score Data (First 5 Rows) ---")
print(metascore_long.head())

sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=metascore_long,
    x='Release Year',
    y='Average_Metacritic_Score',
    hue='Platform',
    marker='o'
)

plt.title(f'Average Metacritic Score Trend for Top 5 Platforms ({START_YEAR}-{END_YEAR})', fontsize=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Average Metacritic Score', fontsize=12)
plt.legend(title='Platform', loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(50, 95) # Set a reasonable y-limit for scores
plt.tight_layout()

plt.savefig('metacritic_score_trend.png')
plt.show()
